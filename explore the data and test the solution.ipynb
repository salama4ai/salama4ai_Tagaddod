```python
def fill_colctr(dfi=dfi):
    """this function is to fill collector_id null values with the collector_id of this collector """
    
    for d in dfi:
        d.fillna(value=(d[d["collector_id"].notna()]["collector_id"][0]), inplace=True)
```

```python
fill_colctr()
(len(df1.collector_id[df1["collector_id"].isna()]), 
 len(df2.collector_id[df2["collector_id"].isna()]), 
 len(df3.collector_id[df3["collector_id"].isna()]), 
 len(df4.collector_id[df4["collector_id"].isna()]))
```
